BUILDING MODEL

LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
lm = LinearRegression()
```

```python
df.columns
```

```python
X = df[['car_ID', 'symboling', 'wheelbase',
       'carlength', 'carwidth', 'carheight', 'curbweight',
        'enginesize', 'boreratio', 'stroke',
       'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg','carbody ordinal','engine ordinal']]
y = df['price']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.40,random_state=101)
```

```python
lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)
```

PREDICTIONS

```python
pred = lm.predict(X_test)
```

```python
plt.scatter(y_test,pred)
```

EVALUATING MODEL

```python
from sklearn import metrics
```

MAE

```python
metrics.mean_absolute_error(y_test,pred)
```

RMSE

```python
np.sqrt(metrics.mean_squared_error(y_test,pred))
```

```python
metrics.r2_score(y_test,pred)
```

CATBOOST REGRESSION 

```python
from catboost import CatBoostRegressor
```

```python
cb = CatBoostRegressor()
```

```python
cb.fit(X_train,y_train)
```

PREDICTIONS

```python
predictionss = cb.predict(X_test)
```

```python
plt.scatter(y_test,predictionss)
```

EVALUATION OF MODEL

```python
np.sqrt(metrics.mean_squared_error(y_test,predictionss))
```

```python
metrics.mean_absolute_error(y_test,predictionss)
```

```python
metrics.r2_score(y_test,predictionss)
```
